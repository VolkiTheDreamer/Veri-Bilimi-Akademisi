# Monty Hall Problemi

```python
import random

def monty_hall_trial(switch: bool) -> bool:    
    all_doors = range(3)
    door_with_car = random.choice(all_doors)
    picked_door = random.choice(all_doors) # oyuncunun ilk seçimi

    # Monty'nin açacağı kapı: oyuncunun seçmediği ve arabanın olmadığı kapılardan biri. İki senaryo var:
    # Oyuncu araba kapısını seçtiyse, Monty diğer iki kapıdan herhangi birini açar. Oyuncu keçi kapısını seçtiyse, Monty diğer keçi kapısını açar.
    possible_opens = [d for d in all_doors if d != picked_door and d != door_with_car] # 1 veya 2 kapı olabilir
    door_monty_opens = random.choice(possible_opens)

    if switch:#fikrini değiştirirse, ilk sçetiği kpaı yerien diğer kapıyı seçerse
        # Kalan kapıya geç: picked_door değil, door_monty_opens değil olan
        # picked_door = next(d for d in all_doors if d != picked_door and d != door_monty_opens)
        picked_door = set(all_doors).difference({picked_door, door_monty_opens}).pop()

    return picked_door == door_with_car #True: kazandı, False: kaybetti

def simulate(n=100_000, seed=42):
    random.seed(seed)
    stay_wins = sum(monty_hall_trial(switch=False) for _ in range(n))
    switch_wins = sum(monty_hall_trial(switch=True) for _ in range(n))
    return stay_wins / n, switch_wins / n

stay_rate, switch_rate = simulate()
print(f"Stay win rate   : {stay_rate:.4f}")
print(f"Switch win rate : {switch_rate:.4f}")
```

```python
import random

def monty_hall_trial_n(n_doors: int, switch: bool) -> bool:
    all_doors = list(range(n_doors))
    door_with_car = random.choice(all_doors)
    picked_door = random.choice(all_doors)

    # Monty'nin açabileceği kapılar: seçilen ve arabanın olmadığı kapılar
    possible_opens = [d for d in all_doors if d != picked_door and d != door_with_car]

    # Monty n-2 kapı açacak (geriye 2 kapı kalsın diye)
    opened = set(random.sample(possible_opens, k=n_doors - 2))

    # Kalan kapılar: opened olmayanlar
    remaining = [d for d in all_doors if d not in opened]

    # remaining tam olarak 2 kapı olmalı: [picked_door, other]
    assert len(remaining) == 2
    other = remaining[0] if remaining[1] == picked_door else remaining[1]

    if switch:
        picked_door = other

    return picked_door == door_with_car

def simulate(n_doors=100, n_trials=200_000, seed=42):
    random.seed(seed)
    stay = sum(monty_hall_trial_n(n_doors, switch=False) for _ in range(n_trials)) / n_trials
    switch = sum(monty_hall_trial_n(n_doors, switch=True) for _ in range(n_trials)) / n_trials
    return stay, switch

how_many=int(input("Kaç kapılı Monty Hall denemesi olsun?: "))
stay, switch = simulate(n_doors=how_many, n_trials=100_000)
print("Stay  :", stay)
print("Switch:", switch)
```
