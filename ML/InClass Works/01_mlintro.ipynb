```python
from sklearn import tree
```

```python

# 1. VERƒ∞ TOPLAMA (DATASET)
# √ñzellikler (Features): [Aƒüƒ±rlƒ±k (gr), Doku]
# Doku Kodu: 0 = P√ºr√ºzs√ºz (Smooth), 1 = P√ºr√ºzl√º (Bumpy)
features = [
    [140, 0], # 140gr, P√ºr√ºzs√ºz -> Elma gibi
    [130, 0], # 130gr, P√ºr√ºzs√ºz -> Elma gibi
    [150, 1], # 150gr, P√ºr√ºzl√º  -> Portakal gibi
    [170, 1]  # 170gr, P√ºr√ºzl√º  -> Portakal gibi
]
```

```python
# Etiketler (Labels): Meyvenin ne olduƒüu
# 0 = Elma (Apple), 1 = Portakal (Orange)
labels = [0, 0, 1, 1]
```

```python
# 2. MODEL SE√áƒ∞Mƒ∞ (Karar Aƒüacƒ± - Decision Tree)
# Karar aƒüacƒ±, veriye bakƒ±p if-else kurallarƒ±nƒ± kendi olu≈üturan bir yapƒ±dƒ±r.
clf = tree.DecisionTreeClassifier()

# 3. Eƒûƒ∞Tƒ∞M (TRAINING)
# .fit() fonksiyonu, veriler (features) ile cevaplarƒ± (labels) e≈üle≈ütirir ve √∂ƒürenir.
# Burasƒ± "Machine Learning"in ger√ßekle≈ütiƒüi yerdir.
clf = clf.fit(features, labels)
```

```python
# 4. TAHMƒ∞N (PREDICTION)
# ≈ûimdi modele daha √∂nce g√∂rmediƒüi bir meyve verelim.
# √ñrn: 160 gram ve P√ºr√ºzl√º (1) bir meyve nedir?
yeni_meyve = [[160, 1]] 

tahmin = clf.predict(yeni_meyve)
```

```python
# Sonucu insan diline √ßevirelim
if tahmin[0] == 0:
    print(f"Girdiƒüimiz meyve: {yeni_meyve}")
    print("Makinenin Tahmini: ELMA üçé")
else:
    print(f"Girdiƒüimiz meyve: {yeni_meyve}")
    print("Makinenin Tahmini: PORTAKAL üçä")
```
