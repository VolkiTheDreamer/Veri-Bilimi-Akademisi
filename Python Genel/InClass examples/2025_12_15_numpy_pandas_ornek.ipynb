```python
def linspacevsarange(n):
    # print(np.arange(n))
    # print(np.linspace(0,n,n))

    for i in range(1,n//2):
      a=np.arange(1,n,i)
      b=np.linspace(1,n,i)
      if len(a)==len(b):
        print("\n***********",i,len(a),a,b,"**********\n")
      else:
        print(i,a,b)


linspacevsarange(100)
# linspacevsarange(105) #bulamayacak
```

```python
df["parantezlimi"]=df["title"].str.contains("\(")
df.query("parantezlimi==False")
```

```python
df["Yıl"]=df["title"].map(lambda x:x.split("(")[1][:-1] if "(" in x else np.nan)
df["#Genres1"]=df["genres"].map(lambda x:len(x.split("|")) if not "(" in x else 0)
df["#Genres2"]=df["genres"].map(lambda x:x.count("|")+1 if not "(" in x else 0)
df.query("parantezlimi==False").head()
```
