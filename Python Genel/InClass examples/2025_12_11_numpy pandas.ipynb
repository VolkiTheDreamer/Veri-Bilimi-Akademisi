```python
df["Yeni1"]=pd.Series([10,20,30,40,50,60,70]) #bi seri ekleyerek
df["Yeni2"]=[-1,-2,-3,-4,-5,-6,-7] #liste ile
df["Yeni3"]=df["Maaslar"]+df["Ucretler"] #iki kolonu toplayarak
df
```

```python
df.Maaslar
```

```python
df["Maaslar"]
```

```python
#aggregate işlemler ile
ortmaas=df.Maaslar.mean()
df["Ort Maas"]=ortmaas
df
```

```python
#list comprehension ile
df["Maas_seviyesi"]=["düşük" if ortmaas>m else "yuksek" for m in df.Maaslar]
df["durum"]=["altta" if i<j else "ustunde" for (i,j) in zip(df.Maaslar,df["Ort Maas"])]
df
```

```python
df.groupby("Maas_seviyesi")["Maaslar"].sum().rename("Seviye_Toplam")
```

```python
seviyetop = df.groupby("Maas_seviyesi")["Maaslar"].sum().rename("Seviye_Toplam").reset_index()
seviyetop
```

```python
df=df.merge(seviyetop)
df
```

```python
#apply, map ve transform gibi metodlar ile
df["maasGrup"]=df.groupby("Maas_seviyesi")["Maaslar"].transform('sum')
df["Oran"]=df["Maaslar"]/df["maasGrup"]
df
```

```python
df["test"]=df.apply(lambda x: x.Yaslar*x.Maaslar,axis=1) #axis=1 dikkat
df
```

```python
df.drop("test",inplace=False,axis=1) #son kolonu silelim
```
