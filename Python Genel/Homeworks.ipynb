# min function implementation

```python
def min_(list_of_numbers):
    """Return the minimum value from a list of numbers."""
    if not list_of_numbers:
        raise ValueError("The list is empty.")
    minimum = list_of_numbers[0] #geçici olarak ilk değeri seçiyoruz
    for number in list_of_numbers:
        if number < minimum:
            minimum = number
    return minimum
    

min_([3, -1, 4, 1, 5, 9, 2, 6, 5])  # Example usage
```

# int function implementation

```python
def int_(s:str)->int:
    """manuel int conversion implementation"""
    negative = False
    result = 0
    start_index = 0

    if len(s) == 0:
        raise ValueError("invalid literal for int() with base 10: ''")

    if s[0] == '-':
        negative = True
        start_index = 1
    elif s[0] == '+':
        start_index = 1

    for i in range(start_index, len(s)):
        char = s[i]
        if char < '0' or char > '9':
            raise ValueError(f"invalid literal for int() with base 10: '{s}'")
        digit = ord(char) - ord('0')
        result = result * 10 + digit

    if negative:
        result = -result

    return result

int("523")  # Example usage
```

# list için sort with return

```python
from forbiddenfruit import curse
def sort_(self,inplace=False):
    """manuel sort implementation"""
    if inplace:
        self.sort()
        return None
    else:
        new_list = sorted(self)
        return new_list
    
curse(list, "sort_", sort_)
```

# list için flatten with return

```python
def flatten_(self,inplace=False):
    """flatten extensionf methdod for lists, flattens nested lists by one level"""
    if inplace:
        i = 0
        while i < len(self):
            if isinstance(self[i], list):
                self[i:i+1] = self[i]
            else:
                i += 1
        return None
    else:
        new_list = []
        for item in self:
            if isinstance(item, list):
                new_list.extend(item)
            else:
                new_list.append(item)
        return new_list    
    
    
curse(list, "flatten_", flatten_)
```

```python
liste= [1,2,[3,4],5,[6,7]]
liste.flatten_()
```

# list için intersect_

```python
def intersect_(self,list2,inplace=True):
    """
        Extension method for list type. It works like sets' counterpart.        
    """    
    if inplace:
        return list(set(self).intersection(set(list2)))
    else:
        temp=self.copy()
        return list(set(temp).intersection(set(list2)))

curse(list, "intersect_", intersect_)
```

```python
liste1=[1,2,3,4,5]
liste2=[4,5,6,7,8]
liste1.intersect_(liste2)
```

# Queue impl.

```python
#Queue implementation
class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if self.is_empty():
            raise IndexError("dequeue from empty queue")
        return self.items.pop(0)

    def peek(self):
        if self.is_empty():
            raise IndexError("peek from empty queue")
        return self.items[0]

    def size(self):
        return len(self.items)
```

```python
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.size()    # returns 3
q.dequeue()  # returns 1
q.size()    # returns 2
```

```python
#Queue Implementation with deque from collections
from collections import deque   
class Queue:
    def __init__(self):
        self.items = deque()

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if self.is_empty():
            raise IndexError("dequeue from empty queue")
        return self.items.popleft()

    def peek(self):
        if self.is_empty():
            raise IndexError("peek from empty queue")
        return self.items[0]

    def size(self):
        return len(self.items)
```

```python
q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.size()    # returns 3
q.dequeue()  # returns 1
q.size()    # returns 2
```

# Stack impl.

```python
# Stack Implementation
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if self.is_empty():
            raise IndexError("pop from empty stack")
        return self.items.pop()

    def peek(self):
        if self.is_empty():
            raise IndexError("peek from empty stack")
        return self.items[-1]

    def size(self):
        return len(self.items)
```

```python
q = Stack()
q.push(1)
q.push(2)
q.push(3)
q.size()    # returns 3
q.pop()     # returns 3
q.size()    # returns 2
```
